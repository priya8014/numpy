```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('customer_shopping_data.csv')
df.head(10)
# df.tail(10)
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
plt.pie(df['Age'])
plt.title('Age Distribution')
plt.show()
```

```python
sns.scatterplot(data=df, x='MonthlyIncome', y='AverageOrderValue', hue='Gender')
plt.title('Income vs. Average Order Value by Gender')
plt.shw()
```

```python
sns.boxplot(data=df, x='SubscriptionStatus', y='CustomerSatisfactionScore')
plt.title('Satisfaction by Subscription')
plt.show()
```

```python
sns.countplot(data=df,y='PreferredCategory',order=df['PreferredCategory'].value_counts().index)
plt.title('MOst preferred Shopping Categories')
plt.show()
```

```python
high_earners = df[(df['MonthlyIncome'] <2000) & (df['CustomerSatisfactionScore'] <=2)]
high_earners.head()
```

```python
df.groupby('Gender')[['MonthlyIncome','CustomerSatisfactionScore']].mean()
```

```python
df['AgeGroup'] = df['Age'].apply(lambda x: 'Youth' if x < 30 else ('Adult' if x < 60 else 'Senior'))
df[['Age', 'AgeGroup']].head()
```

```python
sns.pairplot(df[['Age', 'MonthlyIncome', 'AverageOrderValue', 'Gender']], hue='Gender')
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df=df.rename(columns={'Age':'How Old Are You?'})
```

```python
df
```

```python
df= df.sort_values('How Old Are You?', ascending=False)
```

```python
df= df.sort_values('How Old Are You?', ascending=True)
```

```python
sns.pairplot(df[['How Old Are You?', 'MonthlyIncome', 'AverageOrderValue', 'Gender']], hue='Gender')
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
```

```python
sns.violinplot(data=df, x='AgeGroup', y='AverageOrderValue')
plt.title('Order Value Distribution by Age Group')
```

```python
sns.catplot(data=df, x='SubscriptionStatus', y='CustomerSatisfactionScore', col='AgeGroup', kind='bar')
```

```python


fig, ax = plt.subplots(1, 2, figsize=(12, 5))
sns.boxplot(data=df, x='Gender', y='CustomerSatisfactionScore', ax=ax[0])
ax[0].set_title('Satisfaction by Gender')
sns.boxplot(data=df, x='DeviceType', y='CustomerSatisfactionScore', ax=ax[1])
ax[1].set_title('Satisfaction by Device')
plt.tight_layout()
plt.show()
```
